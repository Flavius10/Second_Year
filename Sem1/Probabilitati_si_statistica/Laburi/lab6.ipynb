```python
from scipy.stats import expon, uniform

# ==============================================================================
# SIMULARE MIXTĂ
# ==============================================================================
def simuleaza_timp_printare(N):
    timpi = []

    # Repetăm simularea de N ori
    for _ in range(N):
        # Pas 1: Alegem imprimanta
        # Generăm un număr uniform u între 0 și 1
        # Dacă u < 0.4 -> Alegem I1 (Prob 0.4)
        # Altfel -> Alegem I2 (Prob 0.6)
        u = uniform.rvs(0, 1)

        if u < 0.4:
            # Imprimanta 1: Exp(1/5) => lambda=1/5 => scale=5 (Media)
            t = expon.rvs(scale=5)
        else:
            # Imprimanta 2: Unif[4, 6] => loc=4, scale=6-4=2
            t = uniform.rvs(loc=4, scale=2)

        timpi.append(t)

    return np.array(timpi)

N_sim = 10000
date_timp = simuleaza_timp_printare(N_sim)

# a) Medie și Deviație Standard
print(f"\na) Medie Timp: {np.mean(date_timp):.4f}")
print(f"   Deviație Std: {np.std(date_timp):.4f}")

# b) Probabilitate P(T < 5)
# Numărăm câte valori sunt < 5
prob_b_sim = np.mean(date_timp < 5)
print(f"b) P(T < 5) estimat: {prob_b_sim:.4f}")

# c) Probabilitate Teoretică
# P(T<5) = P(I1)*P(T1<5 | I1) + P(I2)*P(T2<5 | I2)
# P(T1 < 5) = expon.cdf(5, scale=5)
# P(T2 < 5) = uniform.cdf(5, loc=4, scale=2)

p_t1 = expon.cdf(5, scale=5)      # 1 - e^(-1)
p_t2 = uniform.cdf(5, loc=4, scale=2) # (5-4)/2 = 0.5

prob_b_teo = 0.4 * p_t1 + 0.6 * p_t2
print(f"c) P(T < 5) teoretic: {prob_b_teo:.4f}")
```

```python
# ==============================================================================
# INTEGRARE MONTE CARLO
# ==============================================================================
def g(x):
    return np.exp(-x**2)

# Limitele integralei
a, b = -1, 3
N = 10000

# 1. Generăm puncte uniforme pe intervalul [a, b]
# uniform.rvs(loc=a, scale=b-a)
u = uniform.rvs(loc=a, scale=b-a, size=N)

# 2. Calculăm media valorilor funcției în aceste puncte: E[g(U)]
media_g = np.mean(g(u))

# 3. Integrala ≈ (b - a) * Media
integrala_estimata = (b - a) * media_g

print(f"\n--- Integrare Monte Carlo ---")
print(f"Integrala estimată: {integrala_estimata:.5f}")

# Verificare cu funcția de integrare exactă din scipy (opțional, pt verificare)
from scipy.integrate import quad
val_exacta, _ = quad(g, a, b)
print(f"Valoare exactă:     {val_exacta:.5f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# ==============================================================================
# SUBPUNCTUL i) Simulare și Histogramă
# ==============================================================================
def simuleaza_inaltimi(n):
    # Parametrii distribuției: Media (loc) = 165, Deviația Standard (scale) = 10
    mu, sigma = 165, 10

    # 1. Generăm n valori
    date = norm.rvs(loc=mu, scale=sigma, size=n)

    # 2. Desenăm Histograma
    plt.figure(figsize=(10, 6))

    # range=(130, 210) limitează axa X
    # bins=16 numărul de bare
    # density=True este OBLIGATORIU pentru a putea suprapune graficul teoretic
    plt.hist(date, bins=16, range=(130, 210), density=True,
             color='skyblue', edgecolor='black', alpha=0.7, label='Simulare')

    # 3. Desenăm Densitatea Teoretică (Clopotul lui Gauss)
    # Generăm puncte x pentru linia curbă
    x = np.linspace(130, 210, 1000)
    # Calculăm f(x) teoretic
    y = norm.pdf(x, loc=mu, scale=sigma)

    plt.plot(x, y, 'r-', linewidth=2, label='Teoretic (PDF)')
    plt.title(f"Distribuția Înălțimilor (n={n})")
    plt.legend()
    plt.grid(alpha=0.3)
    plt.show()

    return date

# Rulăm pentru n=5000
date_generate = simuleaza_inaltimi(5000)

# ==============================================================================
# SUBPUNCTUL ii) Statistici (Medie, Deviație, Probabilitate Interval)
# ==============================================================================
# Calculăm valorile simulate
media_sim = np.mean(date_generate)
std_sim = np.std(date_generate)

# Proporția valorilor în [160, 170]
# Numărăm câte sunt în interval și împărțim la total
in_interval = np.sum((date_generate >= 160) & (date_generate <= 170))
prop_sim = in_interval / len(date_generate)

# Valori Teoretice
# Media și deviația sunt chiar parametrii mu și sigma
mu, sigma = 165, 10
# Probabilitatea teoretică P(160 <= X <= 170) = CDF(170) - CDF(160)
prop_teo = norm.cdf(170, loc=mu, scale=sigma) - norm.cdf(160, loc=mu, scale=sigma)

print(f"--- Statistici pentru n=5000 ---")
print(f"Medie:     Sim={media_sim:.2f} | Teo={mu}")
print(f"Deviație:  Sim={std_sim:.2f} | Teo={sigma}")
print(f"P(160-170):Sim={prop_sim:.4f} | Teo={prop_teo:.4f}")






from scipy.stats import randint, bernoulli, binom, hypergeom, geom, uniform, norm, expon

#from itertools import permutations, combinations

#from math import factorial, perm, comb, log, exp, dist
# from random import choice, choices, sample, randint, randrange #random
#from matplotlib.pyplot import hist, bar, plot, xticks, yticks, #axis, legend, grid, show
from numpy import mean, var, std, randint, linspace, log, exp, unique, floor




```
