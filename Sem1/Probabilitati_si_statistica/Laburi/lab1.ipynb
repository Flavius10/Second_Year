```python
from random import sample
from math import factorial
from itertools import permutations

cuvant = 'word'

lista_permutari = list(permutations(cuvant))
print(lista_permutari)

numar_total = factorial(len(cuvant))
print(numar_total)

permutari_aleatoare = sample(cuvant, len(cuvant))
print(permutari_aleatoare)
```

```python
from random import sample
from math import perm, comb
from itertools import permutations, combinations

def aranjamente(multime, k, numar_total=False, aleator=False):
    if numar_total:
        #calculeaza matematic aranjamente de n luate cate k
        return perm(len(multime), k)

    if aleator:
            return sample(multime, k)

    return list(permutations(multime, k))

def combinari(multime, k, numar_total=False, aleator=False):

    if numar_total:
        #calculeaza matematic combinari de n luate cate k
            return comb(len(multime), k)

    if aleator:
        return sample(multime, k)

    return list(combinations(multime, k))

cuvant = 'word'
print("\n--- Teste Aranjamente ---")
print("Lista:", aranjamente(cuvant, 2))
print("Total:", aranjamente(cuvant, 2, numar_total=True))
print("Aleator:", aranjamente(cuvant, 2, aleator=True))

print("\n--- Teste Combinări ---")
print("Lista:", combinari(cuvant, 2))
print("Total:", combinari(cuvant, 2, numar_total=True))
print("Aleator:", combinari(cuvant, 2, aleator=True))
```
