```python
from scipy.stats import norm, uniform, expon
from numpy import mean, std, linspace
from matplotlib.pyplot import show, hist, grid, legend, xticks, plot

n = 1000
m = 165
s = 10

data = norm.rvs(size = n, loc = m, scale = s)

# frecventa relativa este proportia numarului de persoane fata de toate
# density = true, face ca sa fie frecventa relativa
hist(data, bins=16, range=(130, 210), density=True)

# se genereaza multe puncte intr-un anmit interval
x = linspace(130, 210, 1000)

# plot(x, f(x))
plot(x, norm.pdf(x, loc=m, scale=s))

mean(data)
std(data)
print(mean([160 < d < 170 for d in data]))

print(norm.cdf(170, loc=m,scale=s) - norm.cdf(160, loc=m,scale=s))





```

```python
from scipy.stats import norm, uniform, expon

n = 1000
def timpi():
    y = uniform.rvs()

    if y < 0.4:
        t = expon.rvs(loc=0, scale=5)
    else:
        t = uniform.rvs(loc=4, scale=2)

    return t

data = [timpi() for _ in range(n)]
print(mean(data))
print(std(data))

print(mean([d < 5 for d in data]))
print(expon.cdf(5, loc=0,scale=5) * 0.4 + uniform.cdf(5, loc=4,scale=2) * 0.6)
```

```python
from scipy.integrate import quad
from numpy import exp

u = uniform.rvs(loc=-1, scale=4, size=10000)
g = lambda x : exp(-x ** 2)
x = g(u)

print(4 * mean(g(u)))
print(quad(g, -1, 3)[0])
```
