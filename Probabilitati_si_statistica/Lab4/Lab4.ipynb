```python
from scipy.stats import bernoulli
print(bernoulli.rvs(0.5, size = 100));
```

```python
def mers_aleator(n, p):
    lista = bernoulli.rvs(p, size = n)
    poz = 0
    lista_poz = []

    for i in range(n):
        poz += int(2 * lista[i] - 1)
        lista_poz.append(poz)

    return lista_poz

list = mers_aleator(1000, 0.5)
print(list)
#pozitia finala
```

```python
from importlib.metadata import distribution
from random import randrange
from matplotlib.pyplot import bar, hist, grid, show, legend, xticks
from scipy.stats import binom

n = 10
p = 0.5

data = [mers_aleator(n, p)[-1] for _ in range(1000)]

bin_edges = [k - 0.5 for k in range(-n, n + 2)]
hist(data, bins=bin_edges, density=True, rwidth = 0.9, color = 'blue', edgecolor = 'black', alpha = 0.5, label = "Bernoulli")

distribution = dict([(2 * k - n, binom.pmf(k, n, p)) for k in range(n + 1)])
bar(distribution.keys(), distribution.values(), width=0.85,
    color = 'red', edgecolor='black', alpha=0.6, label = "Probabilitati")

xticks(range(-n, n + 2, 2))
legend(loc='upper left')
grid()
show()

# data = binom.rvs(n = 5, p = 6/10, size = 1000)
#
# bin_edges = [k - 0.5 for k in range(7)]
# hist(data, bins=bin_edges, density=True, rwidth=0.9, color='green', edgecolor='black', alpha=0.5, label='frecvente relative')
#
# distribution = dict([(k, binom.pmf(k, n=5, p=6/10)) for k in range(0, 6)])
# bar(distribution.keys(), distribution.values(), width=0.85,
#     color = 'red', edgecolor='black', alpha=0.6, label='probabilitati')
#
# legend(loc='lower left')
# grid()
# show()
```
