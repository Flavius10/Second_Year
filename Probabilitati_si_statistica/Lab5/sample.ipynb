```python

from scipy.stats import uniform, loglaplace
from numpy import mean

Y = uniform.rvs(size = 1000)
mean([(0.25 < y < 0.5) or (0.75 < y < 1) for y in Y])

def my_rnd_gener(x, p, size = 1):
    lista = []

    for _ in range(size):
        Y = uniform.rvs()
        suma = 0

        for i in range(len(p)):
            suma += p[i]
            if Y <= suma:
                lista.append(x[i])
                break

    return lista

l = my_rnd_gener([0, 1, 2, 3], [0.4, 0.2, 0.15, 0.25], 100)
print(l)
```

```python
from matplotlib.pyplot import bar, show, hist, grid, legend, xticks
data = my_rnd_gener([0, 1, 2, 3],[0.4,0.2,0.15, 0.25],size=10000)

bin_edges = [k+0.5 for k in range (-2,4)]

hist (data, bin_edges, density = True, width = 0.9, color = 'pink'
, edgecolor = 'black',
alpha = 0.5)

xticks(range(-2,4,2))
grid()
show()
```

```python
from numpy import log

def my_exp_rnd(a, size = 1):
    lista = []
    for _ in range(size):
        Y = uniform.rvs()
        X = -(1 / a) * log(1 - Y)
        lista.append(X)

    return lista

l = my_exp_rnd(1 / 5, size = 100)
print(l)
```

```python
from matplotlib.pyplot import plot
from scipy.stats import expon
a = 1/5
data = my_exp_rnd(a, size = 100000)
hist(data, bins = 10, density = True, range = (0,30), rwidth = 0.9, color = 'blue', edgecolor = 'black', alpha = 0.5, label = 'generator')

x = range(0,30)
plot(x,expon.pdf(x,loc = 0, scale = 1/a), color = 'red')

xticks(range(0,30,3))
legend(loc = 'upper left')
grid()
show()
```
